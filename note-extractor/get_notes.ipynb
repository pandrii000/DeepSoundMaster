```python
from collections import OrderedDict
```

```python
def get_frequencies_mapping(low=8, high=20000):
    # https://pages.mtu.edu/~suits/NoteFreqCalcs.html
    
    frequencies = {}
    letters = ['A', 'A#', 'B', 'C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#']
    
    f0 = 440
    n = 0
    frequency = f0 * (2 ** (1 / 12)) ** n
    octave = 4
    
    while frequency >= low:
        frequencies[frequency] = letters[n % 12] + str(octave) 
            
        n -= 1
        frequency = f0 * (2 ** (1 / 12)) ** n
        if (n+10) % 12 == 0:
            octave -= 1
            
    n = 1
    frequency = f0 * (2 ** (1 / 12)) ** n
    octave = 4
    
    while frequency <= high:
        frequencies[frequency] = letters[n % 12] + str(octave) 
            
        n += 1
        frequency = f0 * (2 ** (1 / 12)) ** n
        if (n+9) % 12 == 0:
            octave += 1
        
    sorted_frequencies = OrderedDict(sorted(frequencies.items(), key=lambda t: t[0]))
    return sorted_frequencies
```

```python
frequencies = get_frequencies_mapping()
frequencies
```
